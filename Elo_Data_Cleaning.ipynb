```python
# pandas will help us clean the table
import pandas as pd
```

```python
# the csv file we will use for this is assigned to the variable df (dataframe)
df = pd.read_csv(r'nba_elo.csv')
```

```python
# let's run df to see the results
df
```

```python
# We have a bunch of rows we don't need. We are only invested in the Elo and RAPTOR for each team. 
# RAPTOR is a relatively new concept, therefore many years will not have RAPTOR scores.
# We will now perform some data cleaning by using the drop function to get rid of unnecessary columns.
df = df.drop(['neutral', 'playoff', 'carm-elo1_pre', 'carm-elo2_pre', 'carm-elo_prob1', 'carm-elo_prob2', 'carm-elo1_post', 'carm-elo2_post', 'elo1_post', 'elo2_post', 'score1', 'score2', 'quality', 'importance', 'total_rating'], axis = 1)
```

```python
# run the dataframe to see if the columns dropped
df
```

```python
# now we can save this dataframe to a csv file
df.to_csv("total_nba_elo.csv")
```
